```python
def logging_csv(csv_path, timestamp, step, feature_no):
    with open(csv_path, 'a', newline="") as f:
        writer = csv.writer(f)
        writer.writerow(['expert', timestamp, step, 'position', feature_no])
    return
```

```python
csv_path = 'expert_recog.csv'
input_array = []
counter = 0
recog_count = 0

with open('expert_test.csv', encoding='utf-8-sig') as f:
    gesture_data = csv.reader(f)
    # row: list data type for the content in a row
    for row in gesture_data:
        timestamp, x, y, z, step = row[1], row[2], row[3], row[4], row[5]
        input_array.append([x, y, z])
        counter = counter + 1
        if counter == 200:
            print(len(input_array))
            hand_sign_id = gesture_recognizer(input_array)
            print(gesture_recognizer_labels[hand_sign_id])
            feature_no = hand_sign_id + 1
            input_array.clear()
            counter = 0
            logging_csv(csv_path, timestamp, step, feature_no)
```
